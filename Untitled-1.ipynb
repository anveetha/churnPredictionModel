```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot  as plt
import seaborn as sns
from sklearn.model_selection import train_test_split


df = pd.read_csv('train.csv')

#nan_counts_per_column = df.isnull().sum()
#print(nan_counts_per_column)

#if an example has an id, we can assume that the user joined through a referral
df.loc[(df['referral_id'] == 'xxxxxxxx') & (df['joined_through_referral'] == '?'), 'joined_through_referral'] = 'No'
df.loc[(df['referral_id'] != 'xxxxxxxx') & (df['joined_through_referral'] == '?'), 'joined_through_referral'] = 'Yes'
df = df.drop('referral_id', axis = 1)
df = df.drop(['joining_date','last_visit_time'], axis = 1)


df = df.dropna(subset=['preferred_offer_types', 'points_in_wallet', 'region_category'])

df = df.drop(['offer_application_preference', 'security_no','Name','customer_id',], axis = 1 )
df = df.drop(df[df['medium_of_operation'] == '?'].index)
df = df.drop(df[df['gender'] == 'Unknown'].index)
df = df.drop(df[df['avg_frequency_login_days'] == 'Error'].index)

nan_counts_per_column = df.isnull().sum()
print(nan_counts_per_column)
df.head(5)
```

Now that we have a vague idea of what these features look like, lets take a deeper look at what these features contain in each column.

```python
df.dtypes
toManipulate = df.select_dtypes(include = 'object').columns #select features that are string type, and encode/manipulate them accordingly 
print(toManipulate)
```

```python
df.loc[(df['gender'] == 'M'), 'gender'] = 1
df.loc[(df['gender'] == 'F'), 'gender'] = 0
df['gender'] = df['gender'].astype(int)
toManipulate = toManipulate.drop('gender')

df.loc[(df['joined_through_referral'] == 'Yes'), 'joined_through_referral'] = 1
df.loc[(df['joined_through_referral'] == 'No'), 'joined_through_referral'] = 0
toManipulate = toManipulate.drop('joined_through_referral')
df['joined_through_referral'] = df['joined_through_referral'].astype(int)



df.loc[(df['used_special_discount'] == 'Yes'), 'used_special_discount'] = 1
df.loc[(df['used_special_discount'] == 'No'), 'used_special_discount'] = 0
df['used_special_discount'] = df['used_special_discount'].astype(int)
toManipulate = toManipulate.drop('used_special_discount')

df.loc[(df['past_complaint'] == 'Yes'), 'past_complaint'] = 1
df.loc[(df['past_complaint'] == 'No'), 'past_complaint'] = 0
toManipulate = toManipulate.drop('past_complaint')
df['past_complaint'] = df['past_complaint'].astype(int)



ordinal_membership = { #ordinal encoding of membership tiers 
    'No Membership' : 0,
    'Gold Membership':3,
    'Basic Membership' : 1,
    'Silver Membership':2,
    'Premium Membership':4,
    'Platinum Membership':5
}
df['membership_category'] = df['membership_category'].map(ordinal_membership)
toManipulate = toManipulate.drop('membership_category')


ordinal_reviews = { #ordinal encoding of feedback 
    'Poor Website' :3, 
    'No reason specified' : 0, 
    'Poor Product Quality' :2,
    'Too many ads' : 3, 
    'Poor Customer Service' :1, 
    'User Friendly Website' :4,
    'Reasonable Price': 5, 
    'Products always in Stock' : 5,
    'Quality Customer Care' : 5
}
df['feedback'] = df['feedback'].map(ordinal_reviews)
toManipulate = toManipulate.drop('feedback')



ordinal_complaint = { #ordinal encoding of complaints 
    'Solved in Follow-up':2, 
    'Unsolved':3, 
    'Solved' :1, 
    'Not Applicable' : 0,
    'No Information Available':0
}
df['complaint_status'] = df['complaint_status'].map(ordinal_complaint)
toManipulate = toManipulate.drop('complaint_status')


print(toManipulate)
for i in toManipulate: 
    print(print(df[i].unique()))


    
```

```python
df = df.drop(df[df['avg_frequency_login_days'] == 'Error'].index)
df['avg_frequency_login_days'] = df['avg_frequency_login_days'].astype(float)
toManipulate = toManipulate.drop('avg_frequency_login_days')
```

```python
df_encoded = pd.get_dummies(df, columns=['region_category', 'preferred_offer_types', 'medium_of_operation', 'internet_option']) # one-hot encode categories with no sequential ordering
df_encoded.head()
```

```python
df_encoded[df_encoded.select_dtypes('bool').columns] = df_encoded.select_dtypes('bool').astype(int)
df_encoded.head()
```

```python
df_encoded.dtypes #double check column types
```

```python
df = df_encoded #set data frame to encoded data frame
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix


X = df.drop('churn_risk_score', axis = 1)
y = df['churn_risk_score']

Xtrn, Xtest, ytrn, ytest = train_test_split(X, y, test_size=0.33, random_state=42)

from sklearn import tree

decisionTree = tree.DecisionTreeClassifier()
decisionTree.fit(Xtrn, ytrn)
```

```python
predicted = decisionTree.predict(Xtest)
accuracy = accuracy_score(predicted, ytest)
print(accuracy)
```

```python
from sklearn.neural_network import MLPClassifier

neuralNet = MLPClassifier(random_state=1, max_iter=300)
neuralNet.fit(Xtrn, ytrn)
NNpredictions = neuralNet.predict(Xtest)
NNaccuracy = accuracy_score(NNpredictions, ytest)
print(accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
accs = []
for i in range(30):
    knn = KNeighborsClassifier(n_neighbors=i+1)
    knn.fit(Xtrn, ytrn)
    knnPreds = knn.predict(Xtest)
    knnAcc = accuracy_score(knnPreds, ytest)
    accs.append(1-knnAcc)
```

```python
sns.lineplot(accs)
```

```python

taccs = []
for i in range(30):
    knn = KNeighborsClassifier(n_neighbors=i+1)
    knn.fit(Xtrn, ytrn)
    knnPreds = knn.predict(Xtrn)
    knnAcc = accuracy_score(knnPreds, ytrn)
    taccs.append(1-knnAcc)   
```

```python
sns.lineplot(taccs)
```

```python
accuracies = pd.DataFrame({
    'x': list(range(len(accs)) ) * 2,
    'y': accs + taccs,
    'line': ['Line 1'] * len(accs) + ['Line 2'] * len(taccs)
})

sns.lineplot(data=accuracies, x='x', y='y', hue='line')
plt.show()
```
